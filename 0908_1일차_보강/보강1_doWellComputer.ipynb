```python
# 크롤링에 사용할 라이브러리를 설치하고 import 시킨다.
import requests
from bs4 import BeautifulSoup
```

```python
# 타겟 사이트 => http://www.dowellcomputer.com/main.jsp
# requests 모듈의 get() 메소드로 타겟 사이트의 정보를 요청한다.
request = requests.get("http://www.dowellcomputer.com/main.jsp")
print(request) # <Response [200]> => 사이트 정보를 정상적으로 얻어왔다.
```

```python
# 타겟 사이트에서 얻어온 정보 중에서 html 문서만 텍스트로 얻어온다.
html = request.text
print(html)
```

```python
# bs4 모듈의 BeautifulSoup() 함수를 사용해서 html 문서를 파싱(분석)할 준비를 한다.
soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
# html 태그와 html 태그에 지정된 속성(선택자)을 이용한 크롤링
# findAll("태그 이름", {"속성 이름": "속성 값"}) => 태그에 별도의 속성이 없으면 {}안의 내용은 생략할 수 있다.
# 최근 공지사항의 제목은 <b> 태그 내부에 있다.
notices = soup.findAll("b")
# print(notices)
for notice in notices:
    print(notice)
```

```python
# 최근 공지사항의 제목은 class 속성이 tail인 <td> 태그 내부에 있다.
notices = soup.findAll("td", {"class": "tail"})
# print(notices)
for notice in notices:
    print(notice)
```

```python
# 자식 선택자(">") 또는 자손 선택자(" " => 공백 1칸)를 이용한 크롤링
# select("css 선택자")
notices = soup.select("td > a")
# print(notices)
for notice in notices:
    # print(notice)
    # has_attr() 메소드는 인수로 지정된 속성이 있으면 true, 없으면 false를 리턴한다.
    # print(notice.has_attr("href"))
    # get() 메소드는 인수로 지정된 속성의 속성값을 얻어온다.
    # print(notice.get("href"))
    if notice.get("href").find("notice") >= 0:
        print(notice.text)
```
